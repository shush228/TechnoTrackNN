# Building a model

We will use $h_{\theta}(x)$ to denote a linear function.
1. $L = \frac{1}{2m} \Sigma_{i = 1}^{m}(h_{\theta}(x^i) - y^i)^2$


2. $\frac{\delta L}{\delta \theta_j} = \frac{1}{m}(\Sigma_{i = 1}^{m}(h_{\theta}(x^i) - y^i))x_j$


3. $\theta_j = \theta_j - \alpha \frac{\delta L}{\delta \theta_j}$ 

Where $\alpha$ is a learning-rate, $m$ - number of examples. Also all parameters $\theta$ updated simultaneously

```python
class Differentiable:
    def __init__(self):
        pass
    
    def forward(self, **kwargs):
        raise NotImplementedError()
    
    def backward(self, **kwargs):
        raise NotImplementedError()
```

```python
class loss(Differentiable):
    def __init__(self):
        super(loss, self).__init__()
    
    def forward(self, y_pred, y_true): 
        
        loss_value = 0.0
        m = len(y_pred)
        loss_value = 1/(2 * m) * np.sum((y_pred - y_true)**2)
                
        return loss_value
    def backward(self, y_pred, y_true):
        
        partial_grad = np.zeros_like(y_pred-y_true)
        m = len(y_pred)
        
        partial_grad = 1/m * (y_pred - y_true)
        
        return partial_grad
    
```

So, let's implement model of linear regression. We initialize $\theta$ as vector filled with ones.

```python
class linear_regression(Differentiable):
    def __init__(self):
        super(linear_regression, self).__init__()
        self.theta = None
    
    def __call__(self, X):
        
        return self.forward(X)
    
    def backward(self, X):
        
        partial_grad = 0.0
        
        partial_grad = X
        
        return partial_grad
    
    def forward(self, X):
        assert X.ndim == 2, "X should be 2-dimensional: (N of objects, n of features)"
        
        
        if (self.theta is None):
            self.theta = np.ones((1, X.shape[1]))
            
        y_pred = 0.0
        
        y_pred = np.dot(X, self.theta.T)
        
        return y_pred
```

We want to add ones-vector to our dataset $X$ to consider bias. Let implement function, which add this vector and polynomial features.

```python
from sklearn.preprocessing import PolynomialFeatures
def PolyFeatures(X, n):
    poly = PolynomialFeatures(n)
    return poly.fit_transform(X)
```

The next step is impementation of function, which will fit our model. We will add $learning\_rate$ to realize our train loop.

```python
def train_loop(X, y, model, loss_fn, learning_rate, epochs=100):
    loss_history = []
    pbar = tqdm(total=epochs)
    for epoch in range(epochs):
        y_pred = model.forward(X)
        loss_value = loss_fn.forward(y_pred, y)
        grad = np.dot(loss_fn.backward(y_pred, y).T, model.backward(X))
        model.theta = model.theta - learning_rate * grad
        
        loss_history.append(loss_value)
        pbar.update(1)
        pbar.set_postfix({'loss': loss_value})
    pbar.close()
    return loss_history
```

# Fitting the model

```python
learning_rate_range = np.linspace(0.000001, 0.00001, 10) #create an array of learning rate values
min_loss = [] #in this array we will collect minimal loss for every learning rate values
X_train = PolyFeatures(X_train, 1)#add ones-vector to our data
for learning_rate in learning_rate_range:
    obj_fn = loss()
    lr_model = linear_regression()
    loss_history = train_loop(X_train, Y_train, lr_model, obj_fn, learning_rate, epochs=1000)
    min_loss.append(loss_history[-1])
```

```python
print(min_loss)
```

As we can see, the best value of cost function is achieved at learning rate is 0.00001. Let's fit our model again. 

```python
obj_fn = loss()
lr_model = linear_regression()
loss_history = train_loop(X_train, Y_train, lr_model, obj_fn, 0.00001, epochs=1000)
```

Also let's fit our model on validation set

```python
X_val = PolyFeatures(X_val, 1)
obj_fn_val = loss()
lr_model_val = linear_regression()
loss_history_val = train_loop(X_val, Y_val, lr_model, obj_fn, 0.00001, epochs=1000)
```
