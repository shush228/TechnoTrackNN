```python
def one_hot_encode(y):
    y_one_hot = np.zeros(shape = (y.shape[0], 10))
    for i in range(y.shape[0]):
        for j in range(10):
            if y[i] == j:
                y_one_hot[i, j] = 1
            else:
                y_one_hot[i, j] = 0
    return y_one_hot
```

```python
def one_hot_decode(y_one_hot):
    return np.argmax(y_one_hot, axis = 1)
```

```python
y_one_hot_train = one_hot_encode(y_train)
y_one_hot_test = one_hot_encode(y_test)
```

```python
print('Train one_hot shape: {}, Test one_hot shape: {}'.format(y_one_hot_train.shape, y_one_hot_test.shape))
```
